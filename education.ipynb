# ANALYSIS OBJECTIVES

This analysis will look at how different population groups spend their time in terms of getting additional education. I will be looking specifically at:

- Education to get a degree
- Education for personal interest

```python
respondent = pd.read_csv("atusdata/atusresp_2013/atusresp_2013.dat")
```

```python
demographics = pd.read_csv("atusdata/atusrost_2013/atusrost_2013.dat")
```

```python
activities = pd.read_csv("atusdata/atusact_2013/atusact_2013.dat")
```

```python
respondent.info()
```

```python
demographics.info()
```

```python
activities.info()
```

Columns to use in Analysis:

- TUCASEID:  Identifies each Household
- TULINENO: Identifies each person in the household.  Its value is 1 for the person who answered  the survey


From Roster Dataset:
- TEAGE: Respondent Age
- TESEX: Gender
    - 1 Male
    - 2 Female

From Respondent Dataset:

- TEHRUSL1: How many hours do you work a week?
- TEIO1COW: Individual class of worker (government, private, Self employed,etc)
   Valid Entries: 
   - 1 Government federal
   - 2 Government, state
   - 3 Government, local
   - 4 Private, for profit
   - 5 Private, nonprofit
   - 6 Self-employed, incorporated
   - 7 Self-employed, unincorporated 8 Without pay

From Activities Dataset:
- TRCODE: 6 digit code
- TUACTDUR: Duration of activity in minutes

Activity Codes of interest

- 060101: Taking classes for a degree
- 060102:Taking Classes for personal experience

```python
resp_reduced = respondent[['TUCASEID',
                           'TUFINLWGT',
                           'TEHRUSL1',
                           'TEIO1COW',
                           ]]

for invalid in range (-1,-10,-1):
    resp_reduced.replace(invalid, np.nan,inplace = True)
resp_reduced.head()
```

```python
dem_reduced = demographics[['TUCASEID','TULINENO','TEAGE','TESEX']]
#The person that responds the survey is always TULINENO = 1 but the 
#file has demographics for all people in the household. 

dem_reduced = dem_reduced[dem_reduced['TULINENO'] == 1]
del dem_reduced['TULINENO']
for invalid in range (-1,-10,-1):
    dem_reduced.replace(invalid, np.nan,inplace = True)
dem_reduced.head()
```

Merge demographics with respondents by TUCASEID

```python
resp_dem = pd.merge(dem_reduced, resp_reduced, on=['TUCASEID'])
resp_dem.sort_index(inplace=True, by='TUCASEID')
resp_dem.head()
```

```python
codes_desired = [60101, 60102]
act_reduced = activities[activities['TRCODE'].isin(codes_desired)]
act_reduced = act_reduced[['TUCASEID', 'TRCODE', 'TUACTDUR']]
act_reduced.head()
```

Merging respondent/demographics with Activities

```python
all_data = pd.merge(resp_dem, act_reduced, on=['TUCASEID'])
all_data['weighted_minutes'] = all_data['TUFINLWGT'] * all_data['TUACTDUR']
all_data.head()
```

```python
all_data['TEAGE'].plot(kind='hist', alpha = 0.5, title="Histogram for Age")
```

```python
all_data['TESEX'].plot(kind='hist', alpha = 0.5, title="Histogram for Gender")
```

```python
all_data['TEHRUSL1'].plot(kind='hist', 
                          alpha = 0.5,
                          title="Histogram for how many hours people work a week")
```

```python
all_data['TEIO1COW'].plot(kind='hist', 
                          alpha = 0.5,
                          title="Histogram for employee types")
```

```python
def plot_bars(ax,
              legend_labels,
              axis_labels,
              xlabel,
              ylabel,
              title):
    patches, labels = ax.get_legend_handles_labels()
    ax.legend(patches, legend_labels, loc='best')
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xticklabels(axis_labels, rotation=45) 
```

```python
xlabel = "Education Type"
ylabel = "Average Hours"
axis_labels = ['Degree','Personal']
```

## Gender Differences in Time Spent in Education

```python
sum_data=all_data[['TESEX','TUFINLWGT','TRCODE','weighted_minutes']]
sum_data = sum_data.groupby(['TRCODE','TESEX']).sum()
sum_data.columns = ['weight_sum', 'weight_min_sum']
sum_data['weighted_avg'] = (sum_data['weight_min_sum'] / sum_data['weight_sum']/60).round(decimals=1)
ax = sum_data.weighted_avg.unstack(level=1).plot(kind='bar')
title = "Average Hours Spent in Education by Gender and Education Type"
legend_labels=['Male','Female']
plot_bars(ax, legend_labels, axis_labels, xlabel, ylabel, title )
```

## Difference in time Spent in Education by Job Type

```python
sum_data=all_data[['TRCODE','TEIO1COW','TUFINLWGT','weighted_minutes']]
sum_data = sum_data.groupby(['TRCODE','TEIO1COW']).sum()
sum_data.rename(columns={'TUFINLWGT': 'weight_sum','weighted_minutes':'weight_min_sum'},inplace=True)
sum_data['weighted_avg'] = (sum_data['weight_min_sum'] / sum_data['weight_sum']/60).round(decimals=1)
ax = sum_data.weighted_avg.unstack(level=1).plot(kind='bar')
legend_labels=['Fed Gov', 'State Gov', 'Local Gov', 'Priv/Profit', 'Priv/NonProfit', 'Self/Inc','Self/NotInc','No Pay']
title = "Average Hours Spent in Education by Work Type and Education Type"
plot_bars(ax, legend_labels, axis_labels, xlabel, ylabel, title )
```

## Looking at differences by age

```python
sum_data = all_data[['TRCODE','TEAGE','TUFINLWGT','weighted_minutes']]
sum_data['bin']= pd.cut(sum_data['TEAGE'],[20,35,50,65,80,95],retbins=True)[0]
sum_data = sum_data.groupby(['TRCODE','bin']).sum()
sum_data.rename(columns={'TUFINLWGT': 'weight_sum','weighted_minutes':'weight_min_sum'},inplace=True)
sum_data['weighted_avg'] = (sum_data['weight_min_sum'] / sum_data['weight_sum']/60).round(decimals=1)
ax = sum_data.weighted_avg.unstack(level=1).plot(kind='bar')
legend_labels=['20-34', '35-49', '50-64', '65-79', '80-94']
title = "Average Hours Spent in Education by Age Group and Education Type"
plot_bars(ax, legend_labels, axis_labels, xlabel, ylabel, title )
```

## Looking at differences by how many hours they work

```python
sum_data = all_data[['TRCODE','TEHRUSL1','TUFINLWGT','weighted_minutes']]
sum_data['bin']= pd.cut(sum_data['TEHRUSL1'],[20,40,60,80,100],retbins=True)[0]
sum_data = sum_data.groupby(['TRCODE','bin']).sum()
sum_data.rename(columns={'TUFINLWGT': 'weight_sum','weighted_minutes':'weight_min_sum'},inplace=True)
sum_data['weighted_avg'] = (sum_data['weight_min_sum'] / sum_data['weight_sum']/60).round(decimals=1)
ax = sum_data.weighted_avg.unstack(level=1).plot(kind='bar')
legend_labels=['20-39', '40-59', '60-79','80-99']
title = "Average Hours Spent in Education by Number of Hours spent at work and by Education Type"
plot_bars(ax, legend_labels, axis_labels, xlabel, ylabel, title )
```

# Trend Analysis - Years 2003 - 2013

```python
trespondent = pd.read_csv("atusdata/atusresp_0313.dat")
```

```python
tdemographics = pd.read_csv("atusdata/atusrost_0313.dat")
```

```python
tactivities = pd.read_csv("atusdata/atusact_0313.dat")
```

```python
tresp_reduced = trespondent[['TUCASEID',
                           'TUFNWGTP',
                           'TEHRUSL1',
                           'TEIO1COW',
                           ]]

for invalid in range (-1,-10,-1):
    tresp_reduced.replace(invalid, np.nan,inplace = True)
tresp_reduced.head()
```

```python
tdem_reduced = tdemographics[['TUCASEID','TULINENO','TEAGE','TESEX']]
#The person that responds the survey is always TULINENO = 1 but the 
#file has demographics for all people in the household. 

tdem_reduced = tdem_reduced[tdem_reduced['TULINENO'] == 1]
del tdem_reduced['TULINENO']
for invalid in range (-1,-10,-1):
    tdem_reduced.replace(invalid, np.nan,inplace = True)
tdem_reduced.head()
```

```python
tresp_dem = pd.merge(tdem_reduced, tresp_reduced, on=['TUCASEID'])
tresp_dem.sort_index(inplace=True, by='TUCASEID')
tresp_dem.head()
```

```python
codes_desired = [60101, 60102]
tact_reduced = tactivities[tactivities['TRCODEP'].isin(codes_desired)]
tact_reduced = tact_reduced[['TUCASEID', 'TRCODEP', 'TUACTDUR']]
tact_reduced.head()
```

```python
tall_data = pd.merge(tresp_dem, tact_reduced, on=['TUCASEID'])
tall_data['weighted_minutes'] = tall_data['TUFNWGTP'] * tall_data['TUACTDUR']
tall_data['year'] = tall_data['TUCASEID']//10000000000
```

## Exploring how trends have changed during 2003 and 2013

```python
def plot_lines(ax1,ax2,
              legend_labels,
              axis_labels,
              xlabel,
              ylabel,
              title):
    ax1 = fig.add_subplot(211)
    ax2 = fig.add_subplot(212)
    ax1.set_xlabel(xlabel)
    ax1.set_ylabel(xlabel)
    ax2.set_xlabel(ylabel)
    ax2.set_ylabel(ylabel)
    ax1.set_title(title)
    ax2.set_title(title)
    ax1.set_xticklabels(axis_labels, rotation=45) 
    ax2.set_xticklabels(axis_labels, rotation=45) 
```

## Trends in Education by Gender

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
fig.subplots_adjust(hspace=.5)

tsum_data=tall_data[['year','TESEX','TUFNWGTP','TRCODEP','weighted_minutes']]
tsum_data= tsum_data.groupby(['year','TRCODEP','TESEX']).sum()

tsum_data.columns = ['weight_sum', 'weight_min_sum']
tsum_data['weighted_avg'] = (tsum_data['weight_min_sum'] / 
                             tsum_data['weight_sum']/60).round(decimals=1)
del tsum_data['weight_min_sum']
del tsum_data['weight_sum']
tsum_data.reset_index(inplace=True)
tsum_data.set_index(['year','TESEX'],inplace=True)
tsum_temp = tsum_data [tsum_data['TRCODEP'] == 60101]['weighted_avg'].unstack(level=1)
x_labels= tsum_temp.index
ax1.set_title('Trends in Education towards a degree by Gender. Years 2003 - 2013')
ax1.set_ylabel("Average Hours")
ax1.set_xlabel("Years")
ax1.set_xticks(np.arange(0,12))
ax1.set_xticklabels(x_labels)
ax1.plot(tsum_temp)
patches, labels = ax1.get_legend_handles_labels()
legend_labels = ['M','F']
ax1.legend(legend_labels, loc='best')   

ax2 = fig.add_subplot(212)
tsum_temp = tsum_data [tsum_data['TRCODEP'] == 60102]['weighted_avg'].unstack(level=1)
#x_labels= tsum_temp.index
ax2.set_title('Trends in Education for Personal Devleopment by Gender. Years 2003 - 2013')
ax2.set_ylabel("Average Hours")
ax2.set_xlabel("Years")
ax2.set_xticks(np.arange(0,12))
ax2.set_xticklabels(x_labels)
ax2.plot(tsum_temp)
patches, labels = ax2.get_legend_handles_labels()
legend_labels = ['M','F']
ax2.legend(legend_labels, loc='best')        
```
